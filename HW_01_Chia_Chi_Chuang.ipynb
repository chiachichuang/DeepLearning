```python
def read_file(filename):
    line_count = 0
    file_open = open(filename,"r")
    data_array = []
    for line in iter(file_open):
        if "date" in line:
            continue
        line1 = line.strip()
        line_split=line1.split(",")
        data_array.append(line_split)
    return data_array
```

```python
# header
#'"date"', '"Temperature"', '"Humidity"', '"Light"', '"CO2"', '"HumidityRatio"', '"Occupancy"'
filename="occupancy_data/datatraining.txt"
raw_data=read_file(filename)
dataset = np.array(raw_data) # (8143, 8), 1st column:id, 2nd column:date
```

```python
X1=[[datetime.strptime(x.strip('"'), '%Y-%m-%d %H:%M:%S')] for x in dataset[:,1]]
X2=dataset[:,2:7]
Y=dataset[:,7]
X2=X2.astype(np.float)
Y=Y.astype(np.int8)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
# fit a logistic regression model to the data
model = LogisticRegression()
model.fit(X,Y)
print(model)
# make predictions
predicted = model.predict(X)
# summarize the fit of the model
print(metrics.classification_report(Y, predicted))
print(metrics.confusion_matrix(Y, predicted))
```

```python
# Knn- classification
# k-Nearest Neighbor Classification
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X,Y)
print(model)
# make predictions
predicted = model.predict(X)
# summarize the fit of the model
print(metrics.classification_report(Y, predicted))
print(metrics.confusion_matrix(Y, predicted))
```

```python
from keras.models import Sequential
from keras.layers import Dense
# fix random seed for reproducibility
np.random.seed(7) # ????
```

```python
# create model
# input_dim = 5 for 5 input variables
# Specify the neural network model sequentially
model = Sequential()
model.add(Dense(20, input_dim=5, activation='relu')) 
model.add(Dense(10, activation='relu')) 
model.add(Dense(1, activation='sigmoid'))
# Sigmoid to get an output between 0 and 1
```
